```python
import pandas as pd
```

```python
import datetime as dt
```

```python
a = pd.read_csv('purchases.csv')
```

```python
b = pd.read_csv('users.csv')
```

```python
mer = pd.merge(a, b, on='userId',how='inner')
merged = pd.DataFrame(mer)
merged
```

```python
merged['purchaseDate'] = pd.to_datetime(merged['purchaseDate'])
```

```python
merged['month_of_date'] = merged['purchaseDate'].dt.month
```

```python
result = merged.groupby(['month_of_date', 'userId']).agg({'purchaseAmount': 'mean'})
```
